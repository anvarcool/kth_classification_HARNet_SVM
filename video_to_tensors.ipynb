```python
import cv2
import torch
import matplotlib.pyplot as plt
import numpy as np
import os
from tqdm import tqdm
import torchvision as tv
from functools import reduce
import operator
```

```python
data_path = "./KTH_Dataset"
```

```python
os.mkdir(data_path + "_tensors")
```

```python
for directory in (os.listdir(data_path)):
    os.mkdir(os.path.join(data_path + "_tensors", directory))
```

```python
for directory in (os.listdir(data_path)):
    for i, video in enumerate(tqdm(os.listdir(os.path.join(data_path, directory)))):
        cap = cv2.VideoCapture(os.path.join(data_path, directory, video))
        ret, frame = cap.read()
        torch.save(torch.tensor(frame), f"./KTH_Dataset_tensors/{directory}/{video.split('.')[0]}_0.pt")
        i = 0
        while True:
            ret, frame = cap.read()
            if ret:
                i += 1
                torch.save(torch.tensor(frame), f"./KTH_Dataset_tensors/{directory}/{video.split('.')[0]}_{i}.pt")
            else:
                break
        cap.release()
```
